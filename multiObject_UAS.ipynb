## Exploratory Data Analysis (EDA) Dataset


```python
dataset_dir = dataset.location  # Path dari Roboflow download

print(f"  Dataset location: {dataset_dir}")

all_stats = {}

for split in ["train", "valid", "test"]:
    ann_path = os.path.join(dataset_dir, split, "_annotations.coco.json")
    
    if not os.path.exists(ann_path):
        print(f"\n  {split} split tidak ditemukan!")
        continue
    
    with open(ann_path, "r") as f:
        data = json.load(f)
    
    cat_map = {c["id"]: c["name"] for c in data["categories"]}
    cats = Counter([a["category_id"] for a in data["annotations"]])
    
    # Hitung unique source images (pre-augmentation)
    base_names = set()
    for img in data["images"]:
        parts = img["file_name"].split(".rf.")
        base_names.add(parts[0] if len(parts) >= 2 else img["file_name"])
    
    all_stats[split] = {
        "images": len(data["images"]),
        "unique": len(base_names),
        "annotations": len(data["annotations"]),
        "distribution": {cat_map[k]: v for k, v in sorted(cats.items())},
        "categories": data["categories"]
    }
    
    print(f"  {split.upper()} SPLIT")
    print(f"  Total images           : {len(data['images'])}")
    print(f"  Unique source images   : ~{len(base_names)}")
    print(f"  Total annotations (bbox): {len(data['annotations'])}")
    
    print(f"\n  Distribusi Kelas:")
    for cat_id, count in sorted(cats.items()):
        bar = "█" * (count // 30)
        print(f"    {cat_map[cat_id]:15s}: {count:5d}  {bar}")
    
    # Analisis ukuran bbox (hanya untuk train)
    if split == "train":
        sizes = {"small (<32×32)": 0, "medium (32-96)": 0, "large (>96)": 0}
        for ann in data["annotations"]:
            w, h = ann["bbox"][2], ann["bbox"][3]
            area = w * h
            if area < 32 * 32:
                sizes["small (<32×32)"] += 1
            elif area < 96 * 96:
                sizes["medium (32-96)"] += 1
            else:
                sizes["large (>96)"] += 1
        
        total_ann = len(data["annotations"])
        print(f"\n  Distribusi Ukuran BBox (train):")
        for k, v in sizes.items():
            pct = v / total_ann * 100
            bar = "█" * int(pct // 2)
            print(f"    {k:20s}: {v:5d} ({pct:5.1f}%)  {bar}")

# Categories overview
if "train" in all_stats:
    print(f"\n{'─' * 60}")
    print(f"  DAFTAR KELAS ({len(all_stats['train']['categories'])} total)")
    print(f"{'─' * 60}")
    for c in all_stats["train"]["categories"]:
        has_ann = c["name"] in all_stats["train"]["distribution"]
        status = "Done" if has_ann else "⚠️ (0 anotasi — kelas kosong)"
        count = all_stats["train"]["distribution"].get(c["name"], 0)
        print(f"    id={c['id']}: {c['name']:15s} — {count:5d} annotations {status if not has_ann else ''}")

print(f"\n{'=' * 60}")
print(f"  SUMMARY")
print(f"{'=' * 60}")
for split, stats in all_stats.items():
    print(f"    {split:6s}: {stats['images']:4d} images, {stats['annotations']:5d} annotations")
total_img = sum(s["images"] for s in all_stats.values())
total_ann = sum(s["annotations"] for s in all_stats.values())
print(f"    {'TOTAL':6s}: {total_img:4d} images, {total_ann:5d} annotations")
```

## Konfigurasi Hyperparameter


```python
EXPERIMENT_NAME = "exp02_rfdetr_small_e50_bs2"
MODEL_VARIANT   = "RFDETRSmall"

# Training hyperparameters
EPOCHS          = 50
BATCH_SIZE      = 2        # RTX 4050 (6GB VRAM) — batch size 2 untuk hindari OOM
LEARNING_RATE   = 1e-4     # Default RF-DETR
GRAD_ACCUM      = 8        # Effective batch size = BATCH_SIZE * GRAD_ACCUM = 16
DEVICE          = "cuda"

# Paths
DATASET_DIR     = dataset.location
OUTPUT_DIR      = f"./output/{EXPERIMENT_NAME}"

# Inference
CONFIDENCE_THRESHOLD = 0.5

# Print config
print(f"  EXPERIMENT CONFIG: {EXPERIMENT_NAME}")
print(f"  Model           : {MODEL_VARIANT}")
print(f"  Epochs          : {EPOCHS}")
print(f"  Batch size      : {BATCH_SIZE}")
print(f"  Grad accum steps: {GRAD_ACCUM}")
print(f"  Effective batch : {BATCH_SIZE * GRAD_ACCUM}")
print(f"  Learning rate   : {LEARNING_RATE}")
print(f"  Device          : {DEVICE}")
print(f"  Dataset dir     : {DATASET_DIR}")
print(f"  Output dir      : {OUTPUT_DIR}")
```

##  Inisialisasi model RF-DETR Small dengan bobot pre-trained COCO

```python
# OPSI B — default saja (COCO sudah jadi default-nya)
model = RFDETRSmall()
```

## Jalankan proses Fine-tuning (Training)

```python
print(f"  MEMULAI TRAINING: {EXPERIMENT_NAME}")
print(f"  Dataset  : {DATASET_DIR}")
print(f"  Epochs   : {EPOCHS}")
print(f"  Batch    : {BATCH_SIZE} (effective: {BATCH_SIZE * GRAD_ACCUM})")
print(f"  Output   : {OUTPUT_DIR}")

training_start = time.time()

model.train(
    dataset_dir=DATASET_DIR,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    grad_accum_steps=GRAD_ACCUM,
    lr=LEARNING_RATE,
    output_dir=OUTPUT_DIR,
    device=DEVICE
)

training_end = time.time()
training_duration = training_end - training_start

print(f"  TRAINING SELESAI!")
print(f"  Durasi: {training_duration/60:.1f} menit ({training_duration:.0f} detik)")
print(f"  Checkpoint tersimpan di: {OUTPUT_DIR}")
```

## Evaluasi: Inference pada Test Set

```python
# Load kategori dari anotasi
ann_path = os.path.join(DATASET_DIR, "test", "_annotations.coco.json")
with open(ann_path, "r") as f:
    test_data = json.load(f)

cat_map = {c["id"]: c["name"] for c in test_data["categories"]}
class_names = [cat_map[i] for i in sorted(cat_map.keys())]

print(f"  EVALUASI PADA TEST SET")
print(f"  Test images: {len(test_data['images'])}")
print(f"  Classes    : {class_names}")
print(f"  Threshold  : {CONFIDENCE_THRESHOLD}")

# Jalankan prediksi pada semua gambar test
test_dir = os.path.join(DATASET_DIR, "test")
test_images = [f for f in os.listdir(test_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

all_detections = []
print(f"\n  Menjalankan inference pada {len(test_images)} gambar test:")

for i, img_name in enumerate(test_images):
    img_path = os.path.join(test_dir, img_name)
    detections = model.predict(img_path, threshold=CONFIDENCE_THRESHOLD)
    all_detections.append({
        "image": img_name,
        "num_detections": len(detections),
        "detections": detections
    })
    if (i + 1) % 5 == 0 or (i + 1) == len(test_images):
        print(f"    Processed: {i+1}/{len(test_images)}")

# Ringkasan deteksi per kelas
from collections import Counter
class_counts = Counter()
confidence_per_class = {}

for item in all_detections:
    dets = item["detections"]
    if dets.class_id is not None:
        for cls_id, conf in zip(dets.class_id, dets.confidence):
            cls_name = cat_map.get(cls_id, f"unknown_{cls_id}")
            class_counts[cls_name] += 1
            if cls_name not in confidence_per_class:
                confidence_per_class[cls_name] = []
            confidence_per_class[cls_name].append(conf)

print(f"\n{'─' * 60}")
print(f"  RINGKASAN DETEKSI PADA TEST SET")
print(f"{'─' * 60}")
print(f"  {'Kelas':15s} | {'Count':>6s} | {'Avg Conf':>8s} | {'Min':>5s} | {'Max':>5s}")
print(f"  {'─'*15}-+-{'─'*6}-+-{'─'*8}-+-{'─'*5}-+-{'─'*5}")

for cls_name in sorted(class_counts.keys()):
    count = class_counts[cls_name]
    confs = confidence_per_class[cls_name]
    avg_conf = np.mean(confs)
    min_conf = np.min(confs)
    max_conf = np.max(confs)
    print(f"  {cls_name:15s} | {count:6d} | {avg_conf:8.3f} | {min_conf:5.3f} | {max_conf:5.3f}")

total_dets = sum(class_counts.values())
print(f"  {'─'*15}-+-{'─'*6}-+-{'─'*8}-+-{'─'*5}-+-{'─'*5}")
print(f"  {'TOTAL':15s} | {total_dets:6d} |")
```

## Visualisasi Prediksi (Inference Demo)

```python
# Pilih beberapa gambar test secara acak untuk divisualisasikan
NUM_SAMPLES = 6  # Jumlah gambar yang ditampilkan

test_dir = os.path.join(DATASET_DIR, "test")
test_images = [f for f in os.listdir(test_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Ambil sampel (atau semua jika kurang dari NUM_SAMPLES)
samples = random.sample(test_images, min(NUM_SAMPLES, len(test_images)))

# Setup annotator
box_annotator = sv.BoxAnnotator(thickness=2)
label_annotator = sv.LabelAnnotator(text_scale=0.4, text_thickness=1)

# Tentukan grid layout
cols = 3
rows = (len(samples) + cols - 1) // cols
fig, axes = plt.subplots(rows, cols, figsize=(18, 6 * rows))
if rows == 1:
    axes = [axes] if cols == 1 else axes
axes_flat = np.array(axes).flatten()

print(f"Menampilkan {len(samples)} sampel prediksi (threshold={CONFIDENCE_THRESHOLD})\n")

for idx, img_name in enumerate(samples):
    img_path = os.path.join(test_dir, img_name)
    
    # Predict
    detections = model.predict(img_path, threshold=CONFIDENCE_THRESHOLD)
    
    # Load image
    image = np.array(Image.open(img_path))
    
    # Buat labels
    labels = []
    if detections.class_id is not None and len(detections) > 0:
        for cls_id, conf in zip(detections.class_id, detections.confidence):
            cls_name = cat_map.get(cls_id, f"id={cls_id}")
            labels.append(f"{cls_name} {conf:.2f}")
    
    # Annotate
    annotated = box_annotator.annotate(scene=image.copy(), detections=detections)
    annotated = label_annotator.annotate(
        scene=annotated, detections=detections, labels=labels
    )
    
    # Plot
    axes_flat[idx].imshow(annotated)
    axes_flat[idx].set_title(f"{img_name[:30]}... ({len(detections)} deteksi)", fontsize=9)
    axes_flat[idx].axis("off")

# Kosongkan axes yang tidak terpakai
for idx in range(len(samples), len(axes_flat)):
    axes_flat[idx].axis("off")

plt.suptitle(
    f"Hasil Prediksi RF-DETR Small — {EXPERIMENT_NAME}\n"
    f"Threshold: {CONFIDENCE_THRESHOLD}",
    fontsize=14, fontweight="bold"
)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "prediction_samples.png"), dpi=150, bbox_inches="tight")
plt.show()

print(f"\n Gambar tersimpan di: {OUTPUT_DIR}/prediction_samples.png")
```

## Pengujian Data Gambar Eksternal


```python
# Inisialisasi widget upload
uploader = widgets.FileUpload(
    accept='image/*',  # Hanya menerima file format gambar
    multiple=False     # Hanya 1 file sekali upload
)

output_box = widgets.Output()

def on_upload_change(change):
    with output_box:
        clear_output(wait=True)
        if not uploader.value:
            return
            
        # Mengambil file dari uploader
        uploaded_file = uploader.value[0]
        # Mendukung ipywidgets versi lama (dict) maupun versi 8+ (object)
        content = uploaded_file['content'] if isinstance(uploaded_file, dict) else uploaded_file.content
        
        # Simpan gambar sementara
        temp_filename = "temp_inference.jpg"
        with open(temp_filename, "wb") as f:
            f.write(content)
            
        print("Mendeteksi objek pada gambar...")
        
        # Lakukan deteksi (Asumsi 'model' & 'CONFIDENCE_THRESHOLD' sudah ada di memori notebook)
        threshold = globals().get('CONFIDENCE_THRESHOLD', 0.5)
        detections = model.predict(temp_filename, threshold=threshold)
        
        # Baca gambar menggunakan OpenCV
        img = cv2.imread(temp_filename)
        
        # Inisialisasi annotator dari library Supervision
        box_annotator = sv.BoxAnnotator()
        label_annotator = sv.LabelAnnotator()
        
        labels = []
        if detections.class_id is not None:
            for cls_id, conf in zip(detections.class_id, detections.confidence):
                # Asumsi 'class_names' ada di memori dari evaluasi test set sebelumnya
                name = class_names[cls_id] if 'class_names' in globals() else str(cls_id)
                labels.append(f"{name} {conf:.2f}")
                
        # Gambar Bounding Box dan Label ke dalam gambar
        annotated_img = box_annotator.annotate(scene=img, detections=detections)
        annotated_img = label_annotator.annotate(scene=annotated_img, detections=detections, labels=labels)
        
        # Konversi BGR (OpenCV) ke RGB (Matplotlib)
        annotated_img = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)
        
        # Tampilkan Hasil Gambar
        plt.figure(figsize=(10, 10))
        plt.imshow(annotated_img)
        plt.axis('off')
        plt.show()
        
        # Reset uploader agar bisa digunakan berulang kali
        uploader.value = ()

# Pasang observer/event ketika ada file yang diunggah
uploader.observe(on_upload_change, names='value')

print("Silakan upload gambar eksternal Anda:")
display(uploader, output_box)
```

## Pengujian Data Video Eksternal


```python
# --- Konfigurasi ---
CHECKPOINT_PATH = "output/exp02_rfdetr_small_e50_bs2/checkpoint_best_ema.pth"
CONFIDENCE_THRESHOLD = 0.5
LABELS = {
    0: "Attributes",
    1: "dasi",
    2: "sabuk",
    3: "sepatu",
    4: "topiPet",
    5: "wajah"
}

# --- Upload file video ---
print("\nSilakan upload file video (.mp4, .avi, .mov):")
upload_widget = widgets.FileUpload(accept='video/*', multiple=False)
display(upload_widget)

def run_video_inference(change):
    if not upload_widget.value:
        return

    uploaded   = upload_widget.value[0]
    video_name = uploaded['name']
    temp_path  = f"temp_input_{video_name}"
    # ↓ output namanya .avi
    output_name = video_name.rsplit('.', 1)[0] + ".avi"
    output_path = f"output/exp02_rfdetr_small_e50_bs2/video_result_{output_name}"

    with open(temp_path, 'wb') as f:
        f.write(bytes(uploaded['content']))

    # Buka video
    cap = cv2.VideoCapture(temp_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    w   = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h   = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    os.makedirs("output/exp02_rfdetr_small_e50_bs2", exist_ok=True)
    fourcc = cv2.VideoWriter_fourcc(*'XVID')  # ← fix codec
    writer = cv2.VideoWriter(output_path, fourcc, fps, (w, h))

    # Annotator supervision
    box_annotator   = sv.BoxAnnotator(thickness=2)
    label_annotator = sv.LabelAnnotator(text_scale=0.5, text_thickness=1)

    frame_idx = 0
    print(f"\nMemproses video: {video_name}")
    print(f"Resolusi: {w}x{h} | FPS: {fps:.1f} | Total frame: {total_frames}")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_idx += 1
        if frame_idx % 10 == 0:
            print(f"  Frame {frame_idx}/{total_frames} ({100*frame_idx//total_frames}%)", end="\r")

        # Inferensi
        img_pil = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        detections = model.predict(img_pil, threshold=CONFIDENCE_THRESHOLD)

        # Buat label teks
        labels_text = [
            f"{LABELS.get(cls, str(cls))} {conf:.2f}"
            for cls, conf in zip(detections.class_id, detections.confidence)
        ]

        # Annotasi frame
        annotated = box_annotator.annotate(scene=frame.copy(), detections=detections)
        annotated = label_annotator.annotate(scene=annotated, detections=detections, labels=labels_text)
        writer.write(annotated)

    cap.release()
    writer.release()
    os.remove(temp_path)

    print(f"\n✅ Selesai! Video hasil disimpan di: {output_path}")
    print(f"   Total {frame_idx} frame diproses.")

upload_widget.observe(run_video_inference, names='value')
```
